```python
# Title: Web Scraping and Data Extraction using Python

# Description:
# This project focuses on scraping data from websites using Python, extracting useful information, 
# and cleaning the data for further analysis.

# Tools Used: Python, BeautifulSoup, Requests, Pandas, Jupyter Notebook
import requests
```

```python
URL="https://www.scrapethissite.com/pages/simple/"
res=requests.get(URL)
```

```python
res
```

```python
if res.status_code==200:
   print(res.headers)

   with open ("lecture codes/scrap_data.html","w",encoding="utf-8") as f:
        f.write(res.text)
```

```python
from bs4 import BeautifulSoup

with open("lecture codes/scrap_data.html","r",encoding="utf-8") as f:
     html_content=f.read()

soup=BeautifulSoup(html_content,"lxml")
```

```python
all_h3=soup.find_all("h3")

l1=[]

for h3 in all_h3:
     name=h3.get_text(strip=True)
    
     population=h3.find_next("div").select_one("span.country-population").get_text(strip=True)
     l1.append([name,population])
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(l1,columns=["name","population"])
df
```
