```python
# This project scrapes turtle family names and their common names
# from scrapethissite.com using Python and BeautifulSoup.

# The website uses iframes, so special handling is required.
import requests
from bs4 import BeautifulSoup
```

```python
l1=[]
# 1. Load main page

fir_url="https://www.scrapethissite.com/pages/frames/"
res=requests.get(fir_url)
soup=BeautifulSoup(res.text,"lxml")

# 2. Extract iframe URL
iframe=soup.find("iframe")
iframe_url="https://www.scrapethissite.com"+iframe.get("src")

# 3. Collect turtle family links
res_iframe=requests.get(iframe_url)

soup2=BeautifulSoup(res_iframe.text,"lxml")

all_a=soup2.find_all("a")
# 4. Visit each family page
for one_a in all_a :
    page_url=one_a.get("href")

    if page_url==None:
       continue

# 5. Extract family name and common name

    elif "family=" in page_url:
        final_url="https://www.scrapethissite.com"+page_url   #pge_url already contains the /pages/frames/ part

        res_last=requests.get(final_url)
        soup3=BeautifulSoup(res_last.text,"lxml")

        name=soup3.find("h3").get_text(strip=True)
        
        com_name=soup3.find("strong",class_="common-name")
       

        if com_name:
            common_name=com_name.get_text(strip=True)  # more  safe 

        else:
            common_name="not available"

        # print(common_name)

        
        l1.append([name,common_name])
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(l1,columns=["TURTLE NAMES","THEIR COMMAN NAMES"])
df
df.to_csv("turtles_info.csv",index=False)
```
