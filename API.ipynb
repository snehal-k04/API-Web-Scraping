```python
url="https://stephen-king-api.onrender.com/api/books"
res=requests.get(url)
```

```python
data=res.json()
data
```

```python
import pandas as pd
```

```python
df=pd.json_normalize(data["data"])
df.columns
df=df[["Year","Title","Publisher"]]
df
```
